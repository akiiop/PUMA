```python
import numpy as np
import sympy as sy
import mpmath as mp
sy.init_printing() # doctest: +SKIP
```

```python
def BSN(Dimension):
    bs=sy.eye(Dimension)
    for j in range(0,Dimension):
        for l in range(0,Dimension):
            bs[j,l]=sy.exp(2*sy.pi*sy.I*j*l/Dimension)/sy.sqrt(Dimension)
    return bs

def Phase(Dimension,Position,angle):
    phase=sy.eye(Dimension)
    if np.size(Position)==1:
        phase[Position,Position]=sy.exp(sy.I*angle)
        return phase
    else:
        for k in range(np.size(Position)):
            phase[Position[k],Position[k]]=sy.exp(sy.I*angle[k])
        return phase          
```

```python
phi = sy.symbols('phi1:11', real=True)
F3 = BSN(3) 
```

```python
U = Phase(3,[1,2],phi[8:11])*F3*Phase(3,[1,2],phi[6:8])*F3*Phase(3,[1,2],phi[4:6])*F3*Phase(3,[1,2],phi[2:4])*F3*Phase(3,[1,2],phi[0:2])
```

```python
sy.simplify(U)
```

```python
V=F3*Phase(3,1,phi[2])*F3*Phase(3,[1,2],phi[0:2])
```

```python
sy.simplify(V)
```
